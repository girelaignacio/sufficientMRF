```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_table('sim_data.txt', sep=' ')
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from ising_covariates import IsingModel
```

```python
ising_model = IsingModel(n_jobs=-1)
ising_model.fit(X, y, criterion="max")
```

```python
ising_model._coefs
```

```python
Jy = ising_model._coefs['J_y']
hy = ising_model._coefs['h_y']
```
